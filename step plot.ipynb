```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1.5,2.5,3.5,4.5,5.5]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1.5,2.5,3.5,4.5,5.5]

plt.step(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Step Plot Demo")
plt.grid()
plt.show()
```

```python
# add label and color
x=[1,2,3,4,5]
y=[1.5,2.5,3.5,4.5,5.5]

plt.step(x,y,color="r",label="score")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Step Plot Demo")
plt.legend()
plt.grid()
plt.show()
```
