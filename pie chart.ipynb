```python
import matplotlib.pyplot as plt
```

```python
x=[36,34,30,40]
y=["Germany","France","Italy","Brazil"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[36,34,30,40]
y=["Germany","France","Italy","Brazil"]
col=['r','g','y','b']
plt.pie(x,labels=y,colors=col)
plt.show()
```

```python
#explode 
x=[36,34,30,40]
y=["Germany","France","Italy","Brazil"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
#autopct,shadow,radius,label
x=[36,34,30,40]
y=["Germany","France","Italy","Brazil"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1.2,
        labeldistance=1.1,startangle=1.1,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4})
plt.legend(loc=6)
plt.show()
```

```python
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```
