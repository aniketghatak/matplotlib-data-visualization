```python
import matplotlib.pyplot as plt
```

```python
x=[100,200,300,400,500,600,700]
plt.boxplot(x,notch=True)
plt.show()
```

```python
x=[100,200,300,400,500,600,700]
plt.boxplot(x,notch=False)
plt.show()
```

```python
# HorizontalBox plot
x=[100,200,300,400,500,600,700]
plt.boxplot(x,vert=False)
plt.show()
```

```python
#set custom width
x=[100,200,300,400,500,600,700]
plt.boxplot(x,vert=True,widths=0.5)
plt.show()
```

```python
#set custom width,label,patch artist,showmeans
x=[100,200,300,400,500,600,700]
plt.boxplot(x,vert=True,widths=0.5,labels=["Candidate Scores"],patch_artist=True,showmeans=True)
plt.show()
```

```python
x=[100,200,300,400,500,600,700,1200,2100]
plt.boxplot(x,vert=True,widths=0.5,labels=["Candidate Scores"],patch_artist=True,showmeans=True)
plt.show()
```

```python
x=[100,200,300,400,500,600,700,1200,2100]
plt.boxplot(x,vert=True,widths=0.5,labels=["Candidate Scores"],patch_artist=True,showmeans=True,sym="g+")
plt.show()
```

```python
# Join the outliers, colors
x=[100,200,300,400,500,600,700,1200,2100]
plt.boxplot(x,vert=True,widths=0.5,labels=["Candidate Scores"],patch_artist=True,showmeans=True,
           whis=3.5,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

```python
x=[100,200,300,400,500,600,700]
y=[150,250,350,450,550,650,750]
z=[x,y]
plt.boxplot(z,vert=True,widths=0.5,labels=["Candidate Scores A","Candidates Scores B"],
            patch_artist=True,showmeans=True,
           whis=3.5,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```
