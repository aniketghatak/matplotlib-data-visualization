# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[23,22,54,34,73,55]
y=[25,35,22,67,22,52]
color=["r","g","b","y","g","b"]
sizes=[33,22,11,44,55,66]
plt.scatter(x,y,c=color,s=sizes,alpha=0.3,marker="*",)
plt.title("Scatter Plot")
plt.xlabel("Morning")
plt.ylabel("Afternoon")
plt.show()
```

```python
# Color bar
x=[23,22,54,34,73,55]
y=[25,35,22,67,22,52]
color=[11,22,50,33,66,33]
sizes=[133,122,111,144,155,166]
plt.scatter(x,y,c=color,s=sizes,marker="*",cmap="gist_earth")
col_bar=plt.colorbar()
col_bar.set_label("Color Bar")
plt.title("Scatter Plot")
plt.xlabel("Morning")
plt.ylabel("Afternoon")
plt.show()
```
