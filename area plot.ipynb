```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60]
y=[22,31,43,51,64,71]
plt.stackplot(x,y)
plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[22,31,43,51,64,71]
z=[33,24,54,34,56,37]
a=[23,54,64,35,74,32]
plt.stackplot(x,y,z,a)
plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[22,31,43,51,64,71]
z=[33,24,54,34,56,37]
a=[23,54,64,35,74,32]

label=["x","y","z","a"]
plt.stackplot(x,y,z,a,labels=label)
plt.legend()
plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[22,31,43,51,64,71]
z=[33,24,54,34,56,37]
a=[23,54,64,35,74,32]

label=["x","y","z","a"]
plt.stackplot(x,y,z,a,labels=label,colors=["r","g","b"])
plt.legend()
plt.show()
```
