# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=["A","B","C","D","E"]
y=[42,33,47,53,31]
color=["r","g","b","m","y"]
plt.xlabel("Scores",fontsize=12)
plt.ylabel("Candidate",fontsize=12)
plt.title("Examination Result",fontsize=14)
plt.bar(x,y,color=color)
plt.show()
```

```python
x=["A","B","C","D","E"]
y=[42,33,47,53,31]
z=[22,17,39,47,35]
color=["r","g","b","m","y"]
plt.xlabel("Scores",fontsize=12)
plt.ylabel("Candidate",fontsize=12)
plt.title("Examination Result",fontsize=14)
plt.bar(x,y,color="b",label="Section 1")
plt.bar(x,z,color="y",label="Section 2")
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
# make bar graphs side by side
x=["A","B","C","D","E"]
y=[42,33,47,53,31]
z=[22,17,39,47,35]
width=0.2

arr=np.arange(len(x))
arr1=[i+width for i in arr]


plt.xlabel("Scores",fontsize=12)
plt.ylabel("Candidate",fontsize=12)
plt.title("Examination Result",fontsize=14)

plt.bar(arr,y,width,color="b",label="Section 1")
plt.bar(arr1,z,width,color="y",label="Section 2")

plt.xticks(arr+width/2,x)
plt.legend()
plt.show()
```

```python
# Horizontal Bar Graph
x=["A","B","C","D","E"]
y=[42,33,47,53,31]
z=[22,17,39,47,35]
color=["r","g","b","m","y"]

plt.xlabel("Scores",fontsize=12)
plt.ylabel("Candidate",fontsize=12)
plt.title("Examination Result",fontsize=14)

plt.barh(x,y,color="b",label="Section 1")
plt.barh(x,z,color="y",label="Section 2")
plt.legend()
plt.show()
```

```python
# make horizontal bar graphs side by side
x=["A","B","C","D","E"]
y=[42,33,47,53,31]
z=[22,17,39,47,35]
width=0.2

arr=np.arange(len(x))
arr1=[i+width for i in arr]


plt.xlabel("Scores",fontsize=12)
plt.ylabel("Candidate",fontsize=12)
plt.title("Examination Result",fontsize=14)

plt.barh(arr,y,width,color="b",label="Section 1")
plt.barh(arr1,z,width,color="y",label="Section 2")

plt.yticks(arr+width/2,x)
plt.legend()
plt.show()
```
